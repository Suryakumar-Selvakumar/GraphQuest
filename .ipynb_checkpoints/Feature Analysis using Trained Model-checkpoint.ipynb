```python
# import libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
from sklearn.model_selection import train_test_split
import shap
import os
import pickle
```

```python
df = pd.read_csv("data/online_gaming_behavior_dataset_preprocessed.csv")
df.head()
```

```python
# Split dataset

X = df.drop(columns=["EngagementLevel", "PlayerID"])
y = df["EngagementLevel"]

# Obtain train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)
```

```python
# Load Final model
final_model_path = "models/final_xgboost_model.json"
final_model = xgb.Booster()
final_model.load_model(final_model_path)
print("Final Model loaded!")
```

```python
# Compute SHAP Values

shap_values_path = 'data/shap_values.pkl'
dtest = xgb.DMatrix(X_test)

explainer = shap.TreeExplainer(final_model)

if os.path.exists(shap_values_path):
    print("Loading previously saved SHAP values...")
    with open(shap_values_path, 'rb') as file:
        shap_values = pickle.load(file)
else:
    print("Calculating new SHAP values...")
    # Generate SHAP values for the test set
    shap_values = explainer(dtest)

    # Save SHAP values
    with open(shap_values_path, 'wb') as file:
        pickle.dump(shap_values, file)
    print(f"\nSHAP values stored at {shap_values_path}")
```

```python
# Inspect SHAP values

print(f"SHAP values type: {type(shap_values)}")
print(f"SHAP values shape: {shap_values.values.shape}")
```

```python
# We have SHAP values all classes
# Let's split for each class
shap_values_high = shap_values.values[:, :, 2]
shap_values_medium = shap_values.values[:, :, 1]
shap_values_low = shap_values.values[:, :, 0]
```

```python
short_column_names = {
    "Age": "Age",
    "Gender": "Gen",
    "PlayTimeHours": "PTH",
    "InGamePurchases": "IGP",
    "GameDifficulty": "GDif",
    "SessionsPerWeek": "SesWk",
    "AvgSessionDurationMinutes": "ASMin",
    "PlayerLevel": "PyLvl",
    "AchievementsUnlocked": "AsUnl",
    "AvgPlayTimePerWeek": "APWk",
    "isAddicted": "Adctd",
    "Location_Europe": "LoEU",
    "Location_Other": "LoOTH",
    "Location_USA": "LoUSA",
    "Genre_RPG": "RPG",
    "Genre_Simulation": "Sim",
    "Genre_Sports": "Sport",
    "Genre_Strategy": "Strat"
}

# Function to generate legend using abbreviations for SHAP plots
def show_custom_legend(short_column_names):
    fig, ax = plt.subplots(figsize=(12, 1))
    ax.axis("off")

    legend_text = [f"{short_name} – {full_name}" for full_name, short_name in short_column_names.items()]
    
    ax.text(0.99, 0.5, "\n".join(legend_text), ha="right", fontsize=10, fontweight="bold")
    plt.subplots_adjust(right=1.0)
    plt.show()

# Function to create SHAP force plots
def create_force_plot(explainer, shap_values, x_test, idx, class_label):
    shap_values_array = np.array(shap_values[idx])
    feature_values = x_test.iloc[idx].to_numpy()  

    shap.force_plot(
        explainer.expected_value[class_label], 
        shap_values_array,  
        feature_values,
        feature_names=[short_column_names.get(col, col) for col in x_test.columns],
        matplotlib=True,
        show=False
    )
    
    plt.grid(False)
    plt.gca().set_axisbelow(True)
    plt.show()

# Function to generate SHAP force plots for a specific class
def generate_force_plots(explainer, shap_values, x_test, short_column_names, num_samples=3, class_label=2):
    show_custom_legend(short_column_names)

    for idx in range(num_samples):
        create_force_plot(explainer, shap_values, x_test, idx, class_label)
```

## Analyze Features Driving “High” predictions for EngagementLevel

```python
generate_force_plots(explainer, shap_values_high, X_test, short_column_names, num_samples=3, class_label=2)
```

```python
plt.title('Feature Importance', fontsize=14)
shap.summary_plot(shap_values_high, X_test, plot_type="bar", show=True, max_display=10)
```

```python
# Compute the mean absolute SHAP values to determine feature importance
mean_abs_shap = np.abs(shap_values_high).mean(axis=0)
top_features = X_test.columns[np.argsort(mean_abs_shap)[::-1]]  

most_important_feature = top_features[0]  
other_top_features = top_features[1:9]    

fig, axs = plt.subplots(nrows=5, ncols=2, figsize=(16, 24))
plt.subplots_adjust(hspace=0.8, wspace=0.7)  

fig.suptitle(f'SHAP Dependence Plots - {most_important_feature} vs Other Features', fontsize=18, fontweight='bold')

for idx, feature in enumerate(other_top_features):
    row_idx = idx // 2
    col_idx = idx % 2
    
    shap.dependence_plot(feature, shap_values_high, X_test, interaction_index=most_important_feature, ax=axs[row_idx, col_idx], show=False)
    
fig.delaxes(axs[4, 0])
fig.delaxes(axs[4, 1])
plt.tight_layout(rect=[0, 0, 1, 1], pad=3.0)
plt.show()
```
